### Write functions that will (1) convert the answer on each of the personality test questions to a number, as set out in scoring.md, and (2) calculate total scores for each of the subscales (conscientiousness, emotional stability, openness to new experience, agreeableness, extraversion).

```python
def sum_per_subscale(scores_per_person, subscale):
    subscale_sum = sum([eval(scores_per_person[i])[1] for i in range(50) if eval(scores_per_person[i])[0] == subscale])
    return(subscale_sum)

sum_per_subscale(scores_clean.loc[0], 1)

scores_clean['Extraversion'] = scores_clean.apply(lambda row: sum_per_subscale(row, 1), axis=1)
scores_clean['Agreeableness'] = scores_clean.apply(lambda row: sum_per_subscale(row, 2), axis=1)
scores_clean['Conscientiousness'] = scores_clean.apply(lambda row: sum_per_subscale(row, 3), axis=1)
scores_clean['EmotionalStability'] = scores_clean.apply(lambda row: sum_per_subscale(row, 4), axis=1)
scores_clean['Intellect'] = scores_clean.apply(lambda row: sum_per_subscale(row, 5), axis=1)
```

### Create a function/functions that takes in subscale total, and for each subscale, creates a new column called "{subscale_name}_interpretation" (e.g., conscientiousness_interpretation) and applies the labels "low", "medium", or "high" according to a person's score on that subscale. Cut-off scores corresponding to the low, medium and high groups can be found in scoring.md.

```python
Extraversion_low = scores_clean['Extraversion'].mean() - scores_clean['Extraversion'].std()/2
Extraversion_high = scores_clean['Extraversion'].mean() + scores_clean['Extraversion'].std()/2

Agreeableness_low = scores_clean['Agreeableness'].mean() - scores_clean['Agreeableness'].std()/2
Agreeableness_high = scores_clean['Agreeableness'].mean() + scores_clean['Agreeableness'].std()/2

Conscientiousness_low = scores_clean['Conscientiousness'].mean() - scores_clean['Conscientiousness'].std()/2
Conscientiousness_high = scores_clean['Conscientiousness'].mean() + scores_clean['Conscientiousness'].std()/2

Emotional_Stability_low = scores_clean['EmotionalStability'].mean() - scores_clean['EmotionalStability'].std()/2
Emotional_Stability_high = scores_clean['EmotionalStability'].mean() + scores_clean['EmotionalStability'].std()/2

Intellect_low = scores_clean['Intellect'].mean() + scores_clean['Intellect'].std()/2
Intellect_high = scores_clean['Intellect'].mean() + scores_clean['Intellect'].std()/2
```

### Extraversion Low, Mediam and High

```python
Extraversion_interpretation = []

for row in scores_clean['Extraversion']:
    if row < Extraversion_low:
        Extraversion_interpretation.append('Low')
    elif row > Extraversion_high:
        Extraversion_interpretation.append('High')
    else:
        Extraversion_interpretation.append('Medium')
        
scores_clean['Extraversion_interpretation'] = Extraversion_interpretation
```

### Conscientiousness Low, Mediam and high

```python
Conscientiousness_interpretation = []

for row in scores_clean['Conscientiousness']:
    if row < Conscientiousness_low:
        Conscientiousness_interpretation.append('Low')
    elif row > Conscientiousness_high:
        Conscientiousness_interpretation.append('High')
    else:
        Conscientiousness_interpretation.append('Medium')
        
scores_clean['Conscientiousness_interpretation'] = Conscientiousness_interpretation
```

### Emotional_Stability Low, Mediam and high

```python
Emotional_Stability_interpretation = []

for row in scores_clean['EmotionalStability']:
    if row < Emotional_Stability_low:
        Emotional_Stability_interpretation.append('Low')
    elif row > Emotional_Stability_high:
        Emotional_Stability_interpretation.append('High')
    else:
        Emotional_Stability_interpretation.append('Medium')
        
scores_clean['Emotional_Stability_interpretation'] = Emotional_Stability_interpretation
```

### Intellect Low, Medium and High

```python
Intellect_interpretation = []

for row in scores_clean['Intellect']:
    if row < Intellect_low:
        Intellect_interpretation.append('Low')
    elif row > Intellect_high:
        Intellect_interpretation.append('High')
    else:
        Intellect_interpretation.append('Medium')
        
scores_clean['Intellect_interpretation'] = Intellect_interpretation
```
